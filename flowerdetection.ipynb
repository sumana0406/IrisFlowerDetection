PROJECT OVERVIEW:

In this project, I used the Iris Flower dataset to classify three different species of Iris Flowers: Setosa, Versicolor and Virginica. The classification is based on their physical measurements including sepal length, petal length and petal width.


MODEL CHOICE:

I selected the RandomForestClassifier() for this task, as it is a robust machine learning algorithm that performs well on small to medium sized structured datasets. Its ability to handle non-linear relationships and its resistance to overfitting through the use of multiple decision trees makes it particularly suited for this project.


RESULTS:

This project is an example of a classification problem. The model successfully categorizes the flowers into their respective species based on the input features. Key performance metrics such as accuracy, precision, recall, F1 score, and support were used to evaluate the model's performance across each flower species:  



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

data_set = load_iris()
x_data  = data_set.data #flower features
y_data = data_set.target #labels

x_data_train, x_data_test, y_data_train, y_data_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 42)

model = RandomForestClassifier() 
model.fit(x_data_train, y_data_train)

y_predicted = model.predict(x_data_test)

print("Accuracy: ", accuracy_score(y_data_test, y_predicted))
print("Results of Classification: ", classification_report(y_data_test, y_predicted))
```
